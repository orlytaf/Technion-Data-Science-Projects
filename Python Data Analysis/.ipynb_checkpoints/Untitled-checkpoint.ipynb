```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import numpy as np
import pandas as pd
```

```python
churn = pd.read_csv(r'C:\\Users\\User\\Desktop\\python DBA proj 2022\\Python Data Analysis\\churn.csv')
churn.head()
```

```python
churn=churn.drop(['customerID'], axis=1)
churn
```

```python
churn.loc[churn.gender == 'Male', 'gender'] = 1
churn.loc[churn.gender == 'Female', 'gender'] = 0
churn = churn.astype({"gender": 'int64'}) 

churn.Partner = (churn.Partner=='Yes').astype('int')
churn.Dependents = (churn.Dependents=='Yes').astype('int')
churn.PhoneService = (churn.PhoneService=='Yes').astype('int')
churn.MultipleLines = (churn.MultipleLines=='Yes').astype('int')
churn.OnlineSecurity = (churn.OnlineSecurity=='Yes').astype('int')
churn.OnlineBackup = (churn.OnlineBackup=='Yes').astype('int')
churn.DeviceProtection = (churn.DeviceProtection=='Yes').astype('int')
churn.TechSupport = (churn.TechSupport=='Yes').astype('int')
churn.StreamingTV = (churn.StreamingTV=='Yes').astype('int')
churn.StreamingMovies = (churn.StreamingMovies=='Yes').astype('int')
churn.PaperlessBilling = (churn.PaperlessBilling=='Yes').astype('int')
churn.Churn = (churn.Churn=='Yes').astype('int')

churn['TotalCharges'] = pd.to_numeric(churn['TotalCharges'], errors='coerce')

churn
```

```python
churn.isna().mean()
```

```python
churn.info()
```

```python
churn1 = pd.get_dummies(churn)
churn1
```

```python
y=churn.Churn.value_counts(normalize=True)
c = ["#d5815f", "#69b0b1"]
plt.figure(figsize=(5,5))
e = [0,0.1]
y_list = y.tolist()
l_list = ["Active","Churn"]
y.plot(labels = l_list,kind='pie', colors = c, explode = e, autopct='%1.1f%%');
plt.legend(title = "Churn Distribution", loc="upper right")

plt.savefig('churn distribution pie chart.png')

```

```python
y=churn.corr().drop('Churn')
heatmap=sns.heatmap(y,cmap='tab10')
```

```python
churn_corr=churn.corr()['Churn'].drop('Churn').plot(kind='barh')
```

```python
churn.groupby('tenure').agg({'Churn':['count','sum','mean']})
```

```python
churn.loc[(churn['TotalCharges'] <1500),'ChargeGroup'] = '0-1500'
churn.loc[(churn['TotalCharges'] >=1500.) &(churn['TotalCharges'] < 3000),'ChargeGroup'] = '1500-3000'
churn.loc[(churn['TotalCharges'] >=3000.) & (churn['TotalCharges'] < 4500.),'ChargeGroup'] = '3000-4500'
churn.loc[(churn['TotalCharges'] >=4500.) & (churn['TotalCharges'] < 6000.),'ChargeGroup'] = '4500-6000'
churn.loc[(churn['TotalCharges'] >=6000.) & (churn['TotalCharges'] < 7500.),'ChargeGroup'] = '6000-7500'
churn.loc[(churn['TotalCharges'] >= 7500.),'ChargeGroup'] = '7500+'
churn
```

```python
plt.hist(x=churn['TotalCharges'].dropna(), bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
plt.xlabel('TotalCharges Range')
plt.ylabel('Client Amount')
plt.title('Client Amount by Total Charges');
plt.show()
```
